# Data Collection

Data collection and labeling is the backbone of all machine learning applications, from object detection to the Large Language Models, and generative models most everyone is familiar with. 

During the Collection phase for Object detection, pay attention to the following:

- Use high-quality images that are not blurry or too dark.
- Capture objects from different angles and in different lighting conditions.
- Include a variety of backgrounds so the computer doesnâ€™t confuse the object with its surroundings.
- Make sure all object types are represented fairly in the dataset.
- Show objects in different sizes and positions within the image.
- Label each object correctly and consistently.
- Avoid having too many examples of one object and too few of others.

These steps are important because they help the computer learn to recognize objects in real-world situations, not just in the specific images used for training. A well-made dataset leads to a smarter and more reliable model.For our application I am expecting us to want around 1000 pictures as our input set, though we may need more or less depending on our success.


## Data labeling 

There are many tools available to label data, both free and paid. For our purpose I will be pointing students toward Roboflow. Roboflow is an online tool which has a free, student, and paid teir. Students will be able to utilize either the free or student versions, both of which will get them similar featuresets. 

Roboflow does have built in features for training models, though to be able to use these models externally as we would need to, you do need a paid subscription which is roughly $600 a year. If your main intent is to use this tool for dataset generation a few times a year, I highly recommend exporting the dataset out of roboflow, and use it with the sister script in this repo to train your model for free.



### Roboflow: Creating a project


When creating a project in roboflow, make sure to click object detection as the type. Give the project a name, and at this point if you know what objects you will be identifying, add them into the annotation group box. For the free version of roboflow, all datasets are public, be aware of this as you upload photos, and that uploaded photos of people have been taken with their consent.


Once you have a project, you will want to upload your dataset photos.


For each photo you will need to apply a bounding box, as seen above. This box should be as close to the edges of the object as possible, and should match the class we are trying to identify in the image. In the case of multiple objects in one image, make sure to make a box for each individual piece.

Labeling can be spread out across multiple accounts, so students will be able to add labels collaboratively from different computers. 

### Pre/Post Processing



Once all the input data is labeled, select the "New dataset version"



We will always want to Resize our data to 640x640 to match the YOLO model we intend to change. Stretch is generally preffered. Other preprocessors are optional.



Image augmentation is reccomended. This will artifically make the dataset larger by applying transforms or filters to the original dataset. Noise, exposure, rotation, shear, blur, and many of the others are great options for expanding your dataset and increasing robustness of the model in conditions outside of what your original data covered.

Once you hit create, stick with 3x for the maximum version size to get as many augments as possible.

In the versions tab, hit download on your new augmented dataset.



Select YOLOv11, and "Download zip to computer."
